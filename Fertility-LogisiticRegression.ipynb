## Logistic regression. 

```python
kids=pd.read_csv("https://vincentarelbundock.github.io/Rdatasets/csv/AER/Fertility.csv")
kids.drop(columns=['Unnamed: 0'], inplace=True)
kids[['morekids', 'afam', 'hispanic', 'other']]=kids[['morekids', 'afam', 'hispanic', 'other']].eq('yes').mul(1)
kids[['gender1', 'gender2']]=kids[['gender1', 'gender2']].eq('male').mul(1)
def gender (x):
    g=0
    if x[1]==1:
        if x[2]==1:
            g="MM"
            return g
        if x[2]==0:
            g="MF"
            return g
    if x[1]==0:
        if x[2]==1:
            g="MF"
            return g
        if x[2]==0:
            g="FF"
            return g
kids['genders']=kids.apply(gender, axis=1)
#Check for null values

kids.isnull().info()
```

```python
#Min max scaling for continuous data
cont=['age', 'work']
scaler = MinMaxScaler()
norm_kids=kids.copy()
norm_kids[cont] = scaler.fit_transform(norm_kids[cont])
# drop_first=True
norm_kids=pd.get_dummies(norm_kids, columns=['genders'])
norm_kids.drop(columns=['gender1', 'gender2'], inplace=True)
norm_kids
```

### Logistic regression with the statsmodel package. 

```python
formula = "morekids ~ age + C(afam) + C(hispanic) + C(other) + work + C(genders_MF) + C(genders_FF) + C(genders_MM)"
formula2 = "morekids ~ age + C(hispanic) + work + C(genders_MF)"
formula3 = "morekids ~ age + C(hispanic) + work + C(genders)"

model = smf.logit(formula = formula2, data = norm_kids)
lr = model.fit()
print(lr.summary())
```

#### The Pseudo R-squared shows that about 3% of the variance is explained by this model. Hispanic, MF, age, and work are all significant with p < 0.05.

### Results and interpretation of the parameter coefficients for each input variable:
https://www.statsmodels.org/stable/index.html. 

```python
coef=lr.params
np.exp(coef)

# logistic regression coefficients
results = pd.DataFrame(lr.params, columns=["coef"]) 
results["exp_coef"]=np.exp(lr.params)
results
```

#### The coefficients of the logistic regression model can be interpreted as follows:
##### The hispanic coefficient is 0.663401, meaning that if every other factor is held constant and the hispanic variable is 1, the log odd would increase by 0.663401, and therefore the odds of wanting more kids would increase by exp(0.663401) = 1.941385
##### If they have male and female kids, the odds of wanting more kids would increase by 0.745732
##### If the age increases by 1, the odds of wanting more kids would increase by 2.928662
##### If the work hours increase by 1, the odds of wanting more kids would increase by 0.509545

### Model Evaluation

```python
# defining the dependent and independent variables
X= norm_kids[['age', 'hispanic', 'work', 'genders_MF']]
y= norm_kids[["morekids"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)


print("Xk_train shape: ",X_train.shape)
print("Xk_test shape: ",X_test.shape)

print("yk_train shape: ",y_train.shape)
print("yk_test shape: ",y_test.shape)
  
# building the model and fitting the data
log_reg = sm.Logit(y_train, X_train).fit()

# printing the summary table
print(log_reg.summary())
```

```python
yhatTrain = log_reg.predict(X_train)
predictionTrain = list(map(round, yhatTrain))
# performing predictions on the test datdaset
yhat = log_reg.predict(X_test)
prediction = list(map(round, yhat))
 
# # comparing original and predicted values of y
# print('Actual values', list(y_test.values))
# print('Predictions :', prediction)

from sklearn.metrics import (confusion_matrix, accuracy_score)
 
# confusion matrix
cm = confusion_matrix(y_test, prediction)
print ("Confusion Matrix : \n", cm)
 
# accuracy score of the model
print('Train accuracy = ', accuracy_score(y_train, predictionTrain))
print('Test accuracy = ', accuracy_score(y_test, prediction))
```
